# Dropping Columns

## Changing columns - Value,Wage,Release Clause,Joined,Contract Valid Untill,Height and Weight

```python
df1["Value"]=df1['Value'].apply(lambda x:x.lstrip('€'))
```

```python
df1["Wage"]=df1['Wage'].apply(lambda x:str(x).lstrip('€'))
df1["Release Clause"]=df1['Release Clause'].apply(lambda x:str(x).lstrip('€'))
```

```python
df1
```

```python
def str_to_float(x):
    if 'M' in x:
        return float(x.replace('M',''))*1000000
    elif 'K' in x:
        return float(x.replace('K',''))*1000
    else:
        return 0
```

```python
df1['Value']=df1["Value"].apply(str_to_float)
```

```python
df1['Wage']=df1["Wage"].apply(str_to_float)
```

```python
df1['Release Clause']=df1["Release Clause"].apply(str_to_float)
```

```python
df1['Joined']=df1['Joined'].apply(lambda x:str(x)[-4:])
```

```python
df1
```

```python
from datetime import datetime 
```

```python
df1['Contract Valid Until']=pd.to_datetime(df1['Contract Valid Until'])
```

```python
df1.info()
```

```python
def height_conv(x):
    if x is not None:
        return float(str(x).replace("'",'.'))
```

```python
df1['Height']=df1['Height'].apply(height_conv)
```

```python
def weight_conv(x):
    if x is not None:
        return float(str(x).replace("lbs",''))
```

```python
df1['Weight']=df1['Weight'].apply(weight_conv)
```

# Top 5 players in each position 

```python
df1['Position'].unique()
```

```python
def top5_diff_position(arr1):
    y=3
    dataframes=[]
    for x in arr1:
        z='df{}'.format(y)
        z = df1[df1['Position']==x].sort_values(by=['Overall'],ascending=False).head(5)[['Name','Position','Club','Overall']]
        dataframes.append(z)
        y=y+1
    return dataframes
        
```

```python
arr1=df1['Position'].unique()
print((arr1))
top5_diff_position(arr1)
```
